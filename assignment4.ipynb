```python
weather1 = pd.concat([weather_temp,weather_temp1])
weather1
```

```python
weather2 = pd.concat([weather_extra,weather_extra1])
weather2
```

```python
weather = pd.merge(weather1, weather2, on='Date')
weather
```

```python
weather = weather.drop(columns='temp_max')
weather.rename(columns = {"Temp_max" : "temp_max"}, inplace = True)
```

```python
def temp_range(row):
    return (row['temp_max'] - row['temp_min'])
weather['temp_range'] = weather.apply(temp_range, axis=1)
weather.info()
weather
```

```python
weather_18to27 = weather.query("`temp_min` > 65 and `temp_max` < 80")
print(weather_18to27)
```

```python
weather_commoncold = weather.query("`temp_avg` > 64.4")
print(weather_commoncold)
```

```python
weather_common = weather.query("`temp_avg` > 69.8")
print(weather_common)
```

```python
pivot_weather = weather.pivot(index='Date', columns='temp_avg', values='temp_range')
print(pivot_weather)
```

```python
pivot_weather = pivot_weather.reset_index()
melted_weather = pd.melt(pivot_weather, id_vars=['Date'], var_name='temp_avg', value_name='temp_range')
print(melted_weather)
```

```python
group = weather.groupby('Date')
mean = group['temp_avg'].mean()
max = group['temp_max'].max()
min = group['temp_min'].min()
print(mean)
print(max)
print(min)

agg_func = {
    'temp_avg': 'mean',
    'temp_max': 'max',
    'temp_min': 'min'
}
agg_func1 = {
    'temp_max': 'max',
    'temp_min': 'min'
}

mean1 = group.agg(agg_func)
max1 = group.agg(agg_func1)
min1 = group.agg(agg_func1)
print(mean1)
print(max1)
print(min1)

mean2 = group.agg(agg_func).mean()
max2 = group.agg(agg_func1).max()
min2 = group.agg(agg_func1).min()
print(mean2)
print(max2)
print(min2)
```
